```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("f.xlsx")
```

```python
data.info()
```

```python
df = data.dropna()
df.head(10)
```

```python
en_yuksek_uretim_fabrikasi = df.loc[df['UretimeVerilen(KG)'].idxmax()]
print("En fazla üretime çay veren fabrika:")
print("")
print(en_yuksek_uretim_fabrikasi)
```

```python
en_dusuk_uretim_fabrikasi = df.loc[df['UretimeVerilen(KG)'].idxmin()]
print("En az üretime çay veren fabrika:")
print("")
print(en_dusuk_uretim_fabrikasi)
```

```python
en_yuksek_uretim_fabrikasi = df.loc[df['UretimeVerilen(KG)'].idxmax()]
en_dusuk_uretim_fabrikasi = df.loc[df['UretimeVerilen(KG)'].idxmin()]

# Seaborn ile görselleştirme
sns.set(style="whitegrid")
plt.figure(figsize=(40, 6))

# En yüksek üretime sahip fabrikayı görselleştirme
sns.barplot(x='Fabrikalar', y='UretimeVerilen(KG)', data=df, palette='viridis')
plt.scatter(en_yuksek_uretim_fabrikasi.name, en_yuksek_uretim_fabrikasi['UretimeVerilen(KG)'], color='red', s=200, marker='o', label='En Yüksek Üretim')
plt.text(en_yuksek_uretim_fabrikasi.name, en_yuksek_uretim_fabrikasi['UretimeVerilen(KG)'], en_yuksek_uretim_fabrikasi['UretimeVerilen(KG)'], ha='right')
plt.legend()

plt.title('Fabrika Üretim Miktarları')
plt.xlabel('Fabrika')
plt.ylabel('Üretime Verilen (KG)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Fabrikaların Üretim Miktarlarının Dağılımı
sns.barplot(x="Fabrikalar", y="UretilenKuruCay(KG)", data=data)
plt.title("Fabrikaların Ürettiği Kuru Çay Miktarı")
plt.xlabel("Fabrikalar")
plt.ylabel("Üretilen Kuru Çay Miktarı (KG)")
plt.xticks(rotation=50)
plt.show()

# Üretilen toplam kuru çay miktarı
toplam_kuru_cay = data["UretilenKuruCay(KG)"].sum()
print("Toplam Üretilen Kuru Çay Miktarı:", toplam_kuru_cay, "KG")
```
